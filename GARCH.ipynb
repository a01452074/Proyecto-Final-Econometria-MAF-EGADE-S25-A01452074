```python
# =========================
# 1. INSTALAR LIBRERÍA
# =========================
!pip install arch

# =========================
# 2. MONTAR DRIVE
# =========================
from google.colab import drive
drive.mount('/content/drive')

# =========================
# 3. LIBRERÍAS
# =========================
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import glob
import os
from arch import arch_model

# =========================
# 4. PATH ARCHIVOS 5MIN
# =========================
path = '/content/drive/MyDrive/Colab Notebooks/*.xlsx'
files = glob.glob(path)

# Filtrar solo archivos de 5 min
files = [f for f in files if '5min' in f]

print(f"Archivos detectados: {len(files)}")

# =========================
# 5. FUNCIÓN PARA DETECTAR PRECIO
# =========================
def get_price_column(df):
    possible_cols = ['Close', 'Last', 'Price', 'Last Price']
    for col in possible_cols:
        if col in df.columns:
            return col
    return None

# =========================
# 6. LOOP PRINCIPAL
# =========================
results_dict = {}

for file in files:

    try:
        ticker = os.path.basename(file).split('5min')[0]

        df = pd.read_excel(file)
        df.columns = [col.strip() for col in df.columns]

        print(f"\nProcesando: {ticker}")

        # =========================
        # FECHA
        # =========================
        if 'Date' in df.columns and 'Time' in df.columns:
            df['Datetime'] = pd.to_datetime(df['Date'] + ' ' + df['Time'])
        else:
            df['Datetime'] = pd.to_datetime(df.iloc[:,0])

        df = df.sort_values('Datetime')

        # =========================
        # PRECIO
        # =========================
        price_col = get_price_column(df)

        if price_col is None:
            print(f"⚠️ No se encontró precio en {ticker}")
            continue

        price = df[price_col]

        # =========================
        # RETURNS
        # =========================
        returns = np.log(price).diff().dropna()

        if len(returns) < 50:
            print(f"⚠️ Muy pocos datos en {ticker}")
            continue

        # =========================
        # GARCH
        # =========================
        model = arch_model(returns, vol='Garch', p=1, q=1)
        res = model.fit(disp='off')

        results_dict[ticker] = res

        print(f"✅ GARCH listo: {ticker}")

        # =========================
        # GRAFICA VOLATILIDAD
        # =========================
        vol = res.conditional_volatility

        plt.figure()
        vol.plot(title=f'Volatilidad GARCH - {ticker}')
        plt.show()

    except Exception as e:
        print(f"❌ Error en {file}: {e}")

# =========================
# 7. RESUMEN RESULTADOS
# =========================
summary_list = []

for ticker, res in results_dict.items():
    params = res.params

    summary_list.append({
        'Ticker': ticker,
        'omega': params.get('omega', None),
        'alpha': params.get('alpha[1]', None),
        'beta': params.get('beta[1]', None)
    })

summary_df = pd.DataFrame(summary_list)

print("\nResumen GARCH:")
print(summary_df)
```

```python
# Guardar resumen
output_path = '/content/drive/MyDrive/Colab Notebooks/garch_results.csv'
summary_df.to_csv(output_path, index=False)

print(f"\nArchivo guardado en: {output_path}")
```
